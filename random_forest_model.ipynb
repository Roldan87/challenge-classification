```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("csv_output/focus.csv")
target = pd.read_csv("bearing_classes.csv", sep=";")
print(target.columns)
print(target.head())
df2 = df.drop(["experiment_id","bearing_id"], axis=1)
```

```python
y = df2.target.values
X = df2.drop(["target"], axis=1)
print(df2.drop(["target"], axis=1).columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.7, random_state=1,stratify=y)
print(y_train.shape)
print(y_test.shape)
```

```python
forest = RandomForestClassifier(n_estimators=100,random_state=2)
print("done")
forest.fit(X_train, y_train)
```

```python
print(forest.score(X_train, y_train))
print(forest.score(X_test, y_test))
```

```python
cv_results = cross_val_score(forest, X_train, y_train, cv=5)
print(cv_results)
print(np.mean(cv_results))
```

```python
y_predict_test = forest.predict(X_test)
print(y_test)
print(y_predict_test)
```

```python
print(classification_report(y_test, y_predict_test))
```

```python
y_predict = forest.predict(X)
print(y_predict.shape)
```
